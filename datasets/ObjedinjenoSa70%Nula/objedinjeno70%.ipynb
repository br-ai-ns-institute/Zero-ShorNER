```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.DataFrame()
with open('./JNLPBAdump.pickle', 'rb') as f:
    df = df.append(pickle.load(f))
with open('./CDR/CDR_ivi.pkl', 'rb') as f:
    df = df.append(pickle.load(f))
with open('./CHEMDNER/CHEMDNER_ivi.pkl', 'rb') as f:
    df = df.append(pickle.load(f))
with open('./BioRED/BioRED_ivi.pkl', 'rb') as f:
    df = df.append(pickle.load(f))
with open('./NCBI/NCBI_ivi.pkl', 'rb') as f:
    df = df.append(pickle.load(f))
with open('./N2C2/n2c2_ivi.pkl', 'rb') as f:
    df = df.append(pickle.load(f).drop('index', axis=1))
```

```python
df
```

```python
df['labels_sum'] = df.labels.apply(lambda lst: sum(lst))
df['klasa'] = df["class"].apply(lambda lst: ' '.join(lst))
df_hasones = df[df.labels_sum!=0]
df_onlyzeros = df[df.labels_sum==0]
df_new = pd.DataFrame()
for i in df_onlyzeros.klasa.unique():
    df_new = df_new.append(df_onlyzeros[df_onlyzeros.klasa == i].sample(frac = 0.7))
df_new = df_new.append(df_hasones)
df_new.drop(['labels_sum', 'klasa'], inplace=True, axis=1)
df_new = df_new.reset_index()
df_new
```

```python
df_new.to_pickle('./DataSet_ivi_70postoNula.pkl')
```
